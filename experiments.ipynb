```python
from gluoncv.model_zoo import get_model, get_model_list
```

```python
model = get_model("vgg16")
```

```python
model.features
```

```python
base_conv = model.features[0:10]
block1 = model.features[10:15]
block2 = model.features[15:20]
block3 = model.features[20:25]
block4 = model.features[25:30]
```

```python
import mxnet as mx
model.initialize(mx.init.MSRAPrelu())
```

```python
x = mx.nd.ones((32, 3, 224, 224))
base = base_conv(x)
c1 = block1(base)
c2 = block2(c1)
c3 = block3(c2)
c4 = block4(c3)
for i in [c1, c2, c3, c4]:
    print(i.shape)
```
